### legend explaining which line is which plot. Save the plot as a .jpg (“Jpeg”) file.

##### using matplotlib :

```python
# f(x) = [e ^(−x/10)]

x = np.linspace(0,10,100)
y = np.exp(-x/10)*np.sin(x)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)


plt.plot(x,y, 'r')
plt.text(7, 0.75 , r'$f(x) = \epsilon^{-x/10} * sin(x) $')


plt.title(' F(x) ')
plt.xlabel('x' )
plt.ylabel('y')
```

```python
# g(x) = xe^(−x/3)

x = np.linspace(0,10,100)
y = x*np.exp(-x/3)

fig_2 = plt.figure()
ax = fig_2.add_subplot(1, 1, 1)

plt.plot(x,y, 'b')
plt.text(7, 0.8 , r'$ g(x) = x * \epsilon^{-x/3} $')

plt.title(' G(x) ')
plt.xlabel('x')
plt.ylabel('y')
```

##### using seaborn 

```python
import seaborn  as sns
```

```python
# f(x) = [e ^(−x/10)]sin(x)

x = np.linspace(0,10,100)
y = np.exp(-x/10)*np.sin(x)

ax = sns.stripplot(x, y)
ax.set(xlabel ='x',ylabel ='f(x)')
plt.title('F(x) = [e ^(−x/10)]sin(x) ')
```

```python
# g(x) = xe^(−x/3)

x = np.linspace(0,10,100)
y = x*np.exp(-x/3)

ax_1 = sns.stripplot(x, y)
ax_1.set(xlabel ='x',ylabel ='g(x)')
plt.title('G(x) = xe^(−x/3) ')
```
